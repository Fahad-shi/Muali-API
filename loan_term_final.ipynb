```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = df.fillna(df.mean(numeric_only=True))
```

```python
df.Gender = df.Gender.replace({"Male": 1, "Female" : 0})
df.Married = df.Married.replace({"Yes": 1, "No" : 0})
df.Loan_Status = df.Loan_Status.replace({"Y": 1, "N" : 0})
df.Self_Employed = df.Self_Employed.replace({"Yes": 1, "No" : 0})
```

```python
df.info()
```

```python
df['Gender'] = df['Gender'].astype(int)
df['Married'] = df['Married'].astype(int)
df['Self_Employed'] = df['Self_Employed'].astype(int)
df['Self_Employed'] = df['Self_Employed'].astype(int)
df['ApplicantIncome'] = df['ApplicantIncome'].astype(float)
df['Credit_History'] = df['Credit_History'].astype(int)
```

```python
df.info()
```

```python
features  = df.drop(['Loan_ID','Dependents','Property_Area','Education','Loan_Amount_Term','Loan_Status'], axis=1).values
features 
```

```python
target = df['Loan_Amount_Term'].values
target.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(features,target,test_size=0.3,random_state=0)
```

```python
x_train.shape,y_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_train[3:5] = scaler.fit_transform(x_train[3:5])
x_test[3:5] = scaler.transform(x_test[3:5])
```

```python
x_train,x_test
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
ml_model = RandomForestRegressor()
```

```python
ml_model.fit(x_train, y_train)
```

```python
y_pred = ml_model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
score = ml_model.score(x_train, y_train)
print('accuracy_score overall :', score)
print('accuracy_score percent :', round(score*100,2))
```
